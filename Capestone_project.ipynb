```python
#Task2
#To read the project dataframe.csv file
project_df = pd.read_csv("Project_DataFrame.csv")
```

```python
#Task2
# Using for loop to fill the missing values
for i in range(len(project_df)):
    if pd.isnull(project_df.loc[i, 'Cost']):   #Check if value is missing(NaN).
        prev_values = project_df.loc[:i-1, 'Cost'].dropna()  # Take all previous non-null values and compute their mean
        if len(prev_values) > 0:
            running_avg = prev_values.mean()
            project_df.loc[i, 'Cost'] = running_avg  #Replacing the missing values
        else:
            project_df.loc[i, 'Cost'] = 0
```

```python
#Task2
#Saving updated dataframe
project_df.to_csv("Project_Updated_DataFrame.csv", index = False)
```

```python
#Task2
#Display updated project table
print(project_df)
```

```python
#Task4
#Loading all the files after updated
project_df = pd.read_csv("Project_Updated_DataFrame.csv")
employee_df = pd.read_csv("Employee_split_DataFrame.csv")
seniority_level_df = pd.read_csv("Seniority_Level_DataFrame.csv")
```

```python
#Task4
#Joining Project and Employee dataframes using common key i.e "ID" (with the help of merge function)
merge1 = pd.merge(project_df, employee_df, on='ID', how='inner')
```

```python
#Task4
# Joining the merged data frame with Seniority level data frame
Final = pd.merge(merge1, seniority_level_df, on='ID', how='inner')
```

```python
#Task4
#Saving updated single dataframe
Final.to_csv("Final.csv", index=False)
```

```python
#Task5
# Adding Bonus Column(5% of project cost) if the project status is completed, else no bonus (0)
Final['Bonus'] = np.where(Final['Status'] == 'Finished', Final['Cost'] * 0.05, 0)
```

```python
#Task6
#To read the Final with Bonus dataframe.csv file
Final = pd.read_csv("Final_with_bonus.csv")
```

```python
#Task6
#Identifying the failed projects by checking status
mask = Final['Status'].str.contains('fail', case=False, na=False)
```

```python
#Task6
#Demote designation level by 1 for failed projects
Final.loc[mask, 'Designation Level'] = Final.loc[mask, 'Designation Level'] + 1
```

```python
#Task6
#Remove records where Designation Level > 4
Final = Final[Final['Designation Level'] <= 4]  #Here the condition checks final['Designation Level'] <= 4  <-- this creates a boolean series if it is true then retain else it will be remove
```

```python
#Task6
#Saving updated dataframe
Final.to_csv("Final_updated.csv", index=False)
```

```python
#Task7
#To read the updated Final dataframe.csv file
Final = pd.read_csv("Final_updated.csv")
```

```python
#Task7
#Creating logic to add the Mr./Mrs. to the first name
for i in range(len(Final)):
    if Final.loc[i, 'Gender'] == 'M':  #Check the Gender if it is M, return to next statement and add 'Mr.' else add 'Mrs.'
        Final.loc[i, 'First Name'] = 'Mr. ' + Final.loc[i, 'First Name']
    else:
        Final.loc[i, 'First Name'] = 'Mrs. ' + Final.loc[i, 'First Name']
```

```python
#Task7
#Removing the Gender from the Final dataframe after adding prefix
Final.drop('Gender', axis=1, inplace=True)
```

```python
#Task7
#Saving updated dataframe
Final.to_csv("Final_updated_prefix", index=False)
```

```python
#Task8
#Creating logic to promte designation level by 1 for those age is more than 29
for i in range(len(Final)):
    if Final.loc[i, 'Age'] > 29:
        Final.loc[i, 'Designation Level'] = Final.loc[i, 'Designation Level'] - 1
```
